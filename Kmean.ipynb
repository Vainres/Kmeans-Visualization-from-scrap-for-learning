```python
import pygame
from random import randint
import math
from sklearn.cluster import KMeans
def distanceV(p1,p2):
    return math.sqrt((p1[0]-p2[0])**2+(p1[1]-p2[1])**2)

pygame.init()

screen = pygame.display.set_mode((1200,700))

pygame.display.set_caption("kmeans visualization")

BACKGROUND = (214, 214, 214)
BACKGROUND_PANEL=(249,255,230)
WHITE=(255,255,255)
BLACK=(0,0,0)


PanelBorder=(50,50,700,500)
Panel=(55,55,690,490)
PlusButton=(800,50,50,50)
MinusButton=(800,150,50,50)
RunButton=(800,250,150,50)
RandomButton=(800,350,150,50)
AlgorithmButton=(800,450,150,50)
ResetButton=(800,550,150,50)


mouseFont=pygame.font.SysFont('sans',20)
font=pygame.font.SysFont('sans',40)
plusSign=font.render('+',True,WHITE)
minusSign=font.render('-',True,WHITE)
randomSign=font.render('Random',True,WHITE)
resetSign=font.render('Reset',True,WHITE)
algorithmSign=font.render('Algorithm',True,WHITE)
runSign=font.render('Run',True,WHITE)
textK=font.render('K = ',True,WHITE)

running = True
RED = (255,0,0)
GREEN = (0,255,0)
BLUE = (0,0,255)
YELLOW = (147, 153, 35)
PURPLE = (255,0,255)
SKY = (0,255,255)
ORANGE = (255,125,25)
GRAPE = (100,25,125)
GRASS = (55,155,65)

COLORS = [RED,GREEN,BLUE,YELLOW,PURPLE,SKY,ORANGE,GRAPE,GRASS,WHITE]
clock = pygame.time.Clock()
k='0'
points=[]
clusters=[]
pointsLabel=[]
errors=0
run=False
while running:
    
    clock.tick(60)
    screen.fill(BACKGROUND)
    #PANEL
    pygame.draw.rect(screen,BLACK,PanelBorder)
    pygame.draw.rect(screen,WHITE,Panel)
    
    
    #BUTTON BORDER
    pygame.draw.rect(screen,BLACK,PlusButton)
    pygame.draw.rect(screen,BLACK,MinusButton)
    pygame.draw.rect(screen,BLACK,RunButton)
    pygame.draw.rect(screen,BLACK,RandomButton)
    pygame.draw.rect(screen,BLACK,AlgorithmButton)
    pygame.draw.rect(screen,BLACK,ResetButton)
    
    #K VALUE
    kValue=font.render(k,True,WHITE)
    screen.blit(kValue,(960,100))
    
    #BUTTON
    screen.blit(plusSign,(800,55))
    screen.blit(minusSign,(800,155))
    screen.blit(runSign,(800,255))
    screen.blit(randomSign,(800,355))    
    screen.blit(algorithmSign,(800,455))
    screen.blit(resetSign,(800,555))
    screen.blit(textK,(900,100))
    
    #GET mouse position
    mouseX,mouseY=pygame.mouse.get_pos()
    
    #Display panel position
    if 55 < mouseX < 745 and 55 < mouseY < 545:
        textPositionOnPanel=mouseFont.render('('+str(int(mouseX)-55)+','+str(int(mouseY)-55)+')',True,BLACK)
        screen.blit(textPositionOnPanel,(mouseX,mouseY))
        
    
        
    #Event handler
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False
        if event.type == pygame.MOUSEBUTTONDOWN:
        #Plus button            
            if 800 < mouseX < 850 and 50 < mouseY < 100:
                print(mouseX,mouseY)
                if int(k) >=9:
                    k='9'     
                else:
                    k=str(int(k)+1)
        #Minus button
            if 800 < mouseX < 850 and 150 < mouseY < 200:
                print(mouseX,mouseY)
                if int(k) <= 0:
                    k='0'     
                else:
                    k=str(int(k)-1)
        #Random button
            if 800 < mouseX < 950 and 350 < mouseY < 400:
                print(mouseX,mouseY)
                clusters=[]
                for i in range(int(k)):
                    clusters.append([randint(55,745),randint(55,545)])
                    
        #Run button
            if 800 < mouseX < 950 and 250 < mouseY < 300:
                print("run pressed",len(clusters))
                if(len(clusters)>0):
                    run =True
                    for i in range(len(points)):
                        minDistance=distanceV(points[i],clusters[0])
                        minIndex=0
                        for j in range(len(clusters)):
                            distanceBetween=distanceV(points[i],clusters[j])
                            if(distanceBetween<minDistance):
                                minIndex=j
                                minDistance=distanceBetween
                        pointsLabel[i]=minIndex
                    if(len(pointsLabel)>0):
                        for i in range(len(clusters)):
                            sumX=0
                            sumY=0
                            count=0
                            for j in range(len(points)):
                                if(pointsLabel[j]==i):
                                    sumX+=points[j][0]
                                    sumY+=points[j][1]                            
                                    count+=1
                            if(count!=0):
                                clusters[i][0]=int(sumX/count)
                                clusters[i][1]=int(sumY/count)
                    
        #Reset button
            if 800 < mouseX < 950 and 550 < mouseY < 600:
                k='0'
                points=[]
                clusters=[]
                pointsLabel=[]
                errors=0
                run=False
        #A button
            if 800 < mouseX < 950 and 450 < mouseY < 500 and int(k)!=0:
                kmeansModel=KMeans(n_clusters=int(k)).fit(points)
                clusters=kmeansModel.cluster_centers_
                pointsLabel=kmeansModel.predict(points)
        #Panelclick
            if 55 < mouseX < 745 and 55 < mouseY < 545:
                points.append(pygame.mouse.get_pos())
                pointsLabel.append(9)
    if len(clusters)!=0 and len(pointsLabel)!=0:
                        for i in range(len(points)):
                            sumError=0
                            if(pointsLabel[i]!=9):
                                sumError+=distanceV(points[i],clusters[pointsLabel[i]])
                            errors=sumError
    for i in range(len(points)):
        pygame.draw.circle(screen,BLACK,points[i],6)
        pygame.draw.circle(screen,COLORS[pointsLabel[i]],points[i],5)
    
    
    
    ErrorValue=font.render('Error = '+ str(int(errors)),True,WHITE)
    screen.blit(ErrorValue,(1000,100))
    for i in range(len(clusters)):
        pygame.draw.circle(screen,COLORS[i],clusters[i],10)
    pygame.display.flip()

pygame.quit()
```
